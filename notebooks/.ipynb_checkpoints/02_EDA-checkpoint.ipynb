```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings

# Désactiver l'avertissement FutureWarning
warnings.simplefilter(action='ignore', category=FutureWarning)

# Charger les données nettoyées
data = pd.read_csv('../data/data_processed/cleaned_quebec_house_price.csv')

# Remplacer les valeurs infinies par NaN
data.replace([np.inf, -np.inf], np.nan, inplace=True)

# Supprimer les lignes contenant des NaN après traitement
data.dropna(inplace=True)

# Supprimer les valeurs aberrantes pour un meilleur affichage
data = data[data['prix'] < 1e7]  # Exclure les prix supérieurs à 10M
data = data[data['Aire_habitable'] < 10000]  # Exclure les surfaces habitables > 10 000 pi²

# Matrice de corrélation pour les colonnes numériques uniquement
numeric_data = data.select_dtypes(include=[np.number])

plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matrice de Corrélation', fontsize=16)
plt.xticks(fontsize=10, rotation=45, ha='right')
plt.yticks(fontsize=10)
plt.savefig('../reports/figures/matrice_correlation.png')
plt.show()

# Distribution des prix
plt.figure(figsize=(10, 6))
sns.histplot(data['prix'], bins=50, kde=True, color='blue')
plt.title('Distribution des Prix', fontsize=16)
plt.xlabel('Prix ($)', fontsize=12)
plt.ylabel('Fréquence', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.savefig('../reports/figures/distribution_prix_corrected.png')
plt.show()

# Distribution des surfaces habitables
plt.figure(figsize=(10, 6))
sns.histplot(data['Aire_habitable'], bins=50, kde=True, color='green')
plt.title('Distribution des Surfaces Habitables', fontsize=16)
plt.xlabel('Surface Habitable (pi²)', fontsize=12)
plt.ylabel('Fréquence', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.savefig('../reports/figures/distribution_surfaces_corrected.png')
plt.show()

# Relation entre Aire_habitable et prix
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Aire_habitable', y='prix', data=data, color='orange')
plt.title('Relation entre Surface Habitable et Prix', fontsize=16)
plt.xlabel('Surface Habitable (pi²)', fontsize=12)
plt.ylabel('Prix ($)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.savefig('../reports/figures/relation_surface_prix_corrected.png')
plt.show()

# Boxplot des prix par région
plt.figure(figsize=(12, 8))
sns.boxplot(x='region', y='prix', data=data)
plt.title('Prix par Région', fontsize=16)
plt.xlabel('Région', fontsize=12)
plt.ylabel('Prix ($)', fontsize=12)
plt.xticks(rotation=45, fontsize=10, ha='right')
plt.yticks(fontsize=10)
plt.savefig('../reports/figures/prix_par_region_corrected.png')
plt.show()
```
